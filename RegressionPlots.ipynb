```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#seaborn comes with some example datasets

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill', 'tip', tips)
```

```python
sns.lmplot('total_bill', 'tip', tips,
           scatter_kws = {'marker': 'o', 'color': 'indianred'},
           line_kws = {'linewidth': 2, 'color': 'blue'})

#if you add order parameter to lmplot, it will change the fit reg line. 
# (linear line to something else)
```

```python
sns.lmplot('total_bill', 'tip', tips, fit_reg = False)
```

```python
tips.head()
```

```python
tips['tip_pect'] = 100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size', 'tip_pect', tips)
```

```python
sns.lmplot('size', 'tip_pect', tips, x_jitter = .1) #jitter for easy-read
```

```python
sns.lmplot('size', 'tip_pect', tips, x_estimator = np.mean)
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, hue='sex', markers=['x','o'])
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, hue = 'day')
```

```python
#above are linear regression plots
```

```python
#local regression => more specific, statistics subject, seaborn can handle this technique, too.
```

```python
sns.lmplot('total_bill', 'tip_pect', tips, lowess = True, line_kws = {'color': 'black'})
```

```python
sns.regplot('total_bill', 'tip_pect', tips) #regplot is more robust
```

```python
sns.lmplot('total_bill', 'tip_pect', tips) 
```

```python
fig, (axis1, axis2) = plt.subplots(1,2, sharey = True) #row, col
sns.regplot('total_bill', 'tip_pect', tips, ax = axis1)
sns.violinplot(tips['size'], tips['tip_pect'], color = 'indianred', ax = axis2)
```
