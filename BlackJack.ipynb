```python
class Game(object):
    
    
    def __init__(self, suit_, deck_, player, deck):
        self.suit_ = suit_
        self.deck_ = deck_
        Player.player = player
        Deck.deck = deck
    
    def shuffle(self, suit_, deck_):
        random.shuffle(self.suit_)
        random.shuffle(self.deck_)
        
        
    def initializeGame(self, player, deck):
        Player.player.newCard()
        Deck.deck.newCard()
        Player.player.newCard()
        Deck.deck.newCard()
        
        card_counter = 0
        card_cc = 0
        
        for k in Player.player.player_list:
            card_counter += 1
        
        for k in Deck.deck.deck_list:
            card_cc += 1
            
        if card_cc == 2 and card_cc == card_counter:
            Player.player.seeCards()
            Deck.deck.seeCards()
            self.whoWon(Player.player.player_list, Deck.deck.deck_list, Player.player, Deck.deck)
        elif card_cc == 1 and card_counter == 2:
            Deck.deck.newCard()
            Player.player.seeCards()
            Deck.deck.seeCards()
            self.whoWon(Player.player.player_list, Deck.deck.deck_list, Player.player, Deck.deck)
        elif card_counter == 1 and card_cc == 2:
            Player.player.newCard()
            Player.player.seeCards()
            Deck.deck.seeCards()
            self.whoWon(Player.player.player_list, Deck.deck.deck_list, Player.player, Deck.deck)
        else:
            Player.player.newCard()
            Deck.deck.newCard()
            Player.player.seeCards()
            Deck.deck.seeCards()
            self.whoWon(Player.player.player_list, Deck.deck.deck_list, Player.player, Deck.deck)
        
        
    
    def restartOrExitGame(self, player, deck):
        print "Do you want to re-play?"
        replay = int(raw_input('Press 1 for re-play 2 for exit: Decision: '))
        if replay == 1:
            global total_player
            total_player = [[False] * 13, [False] * 13, [False] * 13, [False] * 13]
            Player.player.player_list = []
            Deck.deck.deck_list = []
            self.initializeGame(Player.player,Deck.deck)
        else:
            raise SystemExit("Game is ending...")
        
    def whoWon(self, player_D, deck_D, player, deck):
        dealer_counter = 0
        player_counter = 0
        for iterCard in deck_D:
            if type(iterCard[0]) == int:
                dealer_counter += iterCard[0]
            elif iterCard[0] == 'A':
                dealer_counter += 11
            else:
                dealer_counter += 10
        for iterCard in player_D:
            if type(iterCard[0]) == int:
                player_counter += iterCard[0]
            elif iterCard[0] == 'A':
                player_counter += 11
            else:
                player_counter += 10
        
        if dealer_counter == 21:
            print '--- Dealer won. User lost. Bye! ---'
            self.restartOrExitGame(Player.player, Deck.deck)
        elif player_counter == 21:
            print '--- User won. Dealer lost. Bye! ---'
            self.restartOrExitGame(Player.player, Deck.deck)
        else:
            while True:
                print 'Hit or Stand?:'
                decision = int(raw_input('Hit for 1, Stand for 2: Decision: '))
                if decision == 2:
                    break
                elif decision == 1:
                    print ">>> Player taking new card from deck."
                    Player.player.newCard()
                    Player.player.seeCards()
                    player_counter = 0
                    for iterNew in player_D:
                        if type(iterNew[0]) == int:
                            player_counter += iterNew[0]
                        elif iterCard[0] == 'A':
                            player_counter += 11
                        else:
                            player_counter += 10
                    if player_counter > 21:
                        print '--- Dealer won. User lost. Bye! ---'
                        self.restartOrExitGame(Player.player, Deck.deck)
        
        while dealer_counter <= 16:
            print ">>> Dealer taking new card from deck."
            Deck.deck.newCard()
            Deck.deck.seeCards()
            dealer_counter = 0
            for iterNew in deck_D:
                if type(iterNew[0]) == int:
                    dealer_counter += iterNew[0]
                elif iterCard[0] == 'A':
                    dealer_counter += 11
                else:
                    dealer_counter += 10
            if dealer_counter > 21:
                print '--- User won. Dealer lost. Bye! ---'
                self.restartOrExitGame(Player.player, Deck.deck)
        
        if dealer_counter >= player_counter:
            print ">>> Final Result: "
            Player.player.seeCards()
            Deck.deck.seeCards()
            print '--- Dealer won. User lost. Bye! ---'
            self.restartOrExitGame(Player.player, Deck.deck)
        
        else:
            print ">>> Final Result: "
            Player.player.seeCards()
            Deck.deck.seeCards()
            print '--- User won. Dealer lost. Bye! ---'
            self.restartOrExitGame(Player.player, Deck.deck)
                
    
    
```
