```python
from IPython.display import clear_output
from __future__ import print_function
boardF = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
selections = ['X', 'O']
k_mem = True
g_checker = True
m_checker = False
user_counter = 0

def display_board(board):
    counter = 0
    for k in board:
        if counter % 3 == 0:
            print('')
            print('  ___   ___   ___')
            print(' | {x} |'.format(x = k), end = '')
            counter += 1
        else:
            print(' | {x} |'.format(x = k), end = '')
            counter += 1
```

```python
def player_input():
    
    a_checker = True
    while a_checker:
        display_board(boardF)
        print('')
        if user_counter % 2 == 0:
            pl_input = 'X'
        else:
            pl_input = 'O'
        pl_place = raw_input('Which position 1-9?:')
        print("Marker: {x}, Position: {y} ".format(x = pl_input, y = pl_place))
        print('')
        pl_input = str(pl_input)
        pl_place = int(pl_place)
        if marker_checker(pl_input):
            print('Error, enter proper marker[X&O].')
            print('')
        if range_checker(pl_place):
            print('Error, enter proper place[1-9]')
            print('')
        else:
            place_marker(boardF, pl_input, pl_place)
            a_checker = False
```

```python
def marker_checker(marker):
    if marker == 'X' or marker == 'O':
        return False
    else:
        return True
```

```python
def range_checker(range):
    if range in xrange(1,10):
        return False
    else:
        return True
```

```python
def place_marker(board, marker, position):
    global user_counter
    checker = False
    for k in board:
        if k == ' ':
            checker = True
            
    if place_checker(board,marker,position) and checker:
        board[position-1] = marker
        user_counter += 1
    else:
        print('Slot is full!')
        win_check(board, marker)
        print('')
        player_input()
```

```python
def place_checker(board, marker,position):
    if(board[position-1] == 'X' or board[position-1] == 'O'):
        return False
    else:
        return True
```

```python
def win_check(board, marker):
    global g_checker
    global m_checker
    global k_mem
    if condition_control(board, marker):
        display_board(boardF)
        print('')
        print('Player with marker: {x} won the game!'.format(x = marker))
        print('')
        m_checker = True
        g_checker = False
        k_mem = False
        return
    else:
        board_finisher = True
        for k in board:
            if k == ' ':
                board_finisher = False
        
        if board_finisher and k_mem:
            print('Game tie!')
            m_checker = True
            g_checker = False
            return
```

```python
def condition_control(board, marker):
    
    if (marker == ' '):
        return false
    else:
        p1 = board[1] == marker and board[2] == marker and board[3] == marker
        p2 = board[3] == marker and board[4] == marker and board[5] == marker
        p3 = board[6] == marker and board[7] == marker and board[8] == marker
        p4 = board[0] == marker and board[3] == marker and board[6] == marker
        p5 = board[1] == marker and board[4] == marker and board[7] == marker
        p6 = board[2] == marker and board[5] == marker and board[8] == marker
        p7 = board[0] == marker and board[4] == marker and board[8] == marker
        p8 = board[2] == marker and board[4] == marker and board[6] == marker
        
        return p1 or p2 or p3 or p4 or p5 or p6 or p7 or p8
```

```python
print ('Welcome to Tic Tac Toe!')

while True:
    global boardF
    while g_checker:
        win_check(boardF, selections[0])
        win_check(boardF, selections[1])
        if m_checker == False:
            player_input()
        else:
            break
    
    print('Want to play again?')
    alpha = raw_input('Enter "1" to play again: ')
    alpha = int(alpha)
    if alpha == 1:
        g_checker = True
        m_checker = False
        boardF = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
    else:
        print('Bye!')
        break;
```
