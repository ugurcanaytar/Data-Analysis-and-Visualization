```python
dx,dy = np.meshgrid(points,points)
```

```python
points2 = np.arange(-5,5,1)
```

```python
dxx,dyy = np.meshgrid(points2, points2)
```

```python
dxx
```

```python
dyy
```

```python
k = (np.cos(dxx) + np.cos(dyy))
```

```python
plt.imshow(k)
plt.colorbar()
plt.title('UA Training Graph')
```

```python
# numpy where

A = np.array([1,2,3,4])
B = np.array([100,200,300,400])
```

```python
condition = np.array([True,True,False,False])
```

```python
answer = [(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)] #manual
```

```python
answer
```

```python
answer2 = np.where(condition,A,B) #if condition met, choose A otw choose B => automatic
```

```python
answer2
```
